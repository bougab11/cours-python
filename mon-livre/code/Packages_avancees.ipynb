## FuncAnimation
**Funcanim** est un module de Matplotlib qui permet de créer des animations à partir de graphiques. Il offre un contrôle précis sur chaque image de l'animation, permettant de visualiser des processus dynamiques et des changements au fil du temps.

** Fonction `FuncAnimation`**

* La fonction principale pour créer des animations.
* Prend en arguments :
* La figure (`fig`) sur laquelle l'animation sera dessinée.
* Une fonction (`func`) qui met à jour le graphique à chaque image.
* Les images (`frames`) qui définissent les données ou les paramètres pour chaque image.
* L'intervalle (`interval`) entre les images en millisecondes.

**Exemple de base:**

```python
%matplotlib ipympl

plt.close()
from matplotlib.animation import FuncAnimation

fig, ax2 = plt.subplots()
xdata, ydata = [], []
line, = ax2.plot([], [])  # Ajout d'une couleur pour meilleure visibilité

# Définir les limites des axes
ax2.set_xlim(0, 2 * np.pi)
ax2.set_ylim(-1.5, 1.5)

def init():  # Fonction d'initialisation pour établir le fond de l'animation
    line.set_data([], [])
    return line,

def update(frame):
    xdata.append(frame)
    ydata.append(np.sin(frame))
    line.set_data(xdata, ydata)
    return line,

# Créer l'animation
ani = FuncAnimation(fig, update, frames=np.linspace(0,  2 * np.pi, 128),
                    init_func=init, interval=5, repeat = False)  # Interval ajusté pour une visualisation plus claire

plt.show()
```

Cet exemple crée une animation d'un point se déplaçant le long d'une sinusoïde. La fonction `update` est appelée à chaque image pour mettre à jour les données du graphique.

###  Techniques d'animation

* **Mise à jour des données:** Modifier les valeurs des données à chaque image pour visualiser des changements au fil du temps.
* **Modification des propriétés des artistes:** Changer les couleurs, les tailles, les positions des éléments du graphique.
* **Ajout et suppression d'artistes:** Faire apparaître ou disparaître des éléments pour visualiser des processus dynamiques.

### Applications de Funcanim

* Visualisation de simulations scientifiques.
* Animation de séries temporelles.
* Création de visualisations interactives.
* Exploration de l'espace des paramètres.

**Funcanim est un outil puissant pour créer des visualisations dynamiques et interactives avec Matplotlib. En utilisant des techniques d'animation, vous pouvez donner vie à vos données et mieux comprendre les processus sous-jacents.**


# Numba : Niveau Intermédiaire 

Cette section explore les fonctionnalités avancées de Numba pour optimiser encore davantage le code Python numérique et exploiter les capacités de calcul parallèle.


## Types et signatures de fonctions

* **Annotations de types:** Déclarer explicitement les types des arguments et des valeurs de retour pour améliorer l'efficacité de la compilation.


```python
from numba import int32, float64, njit, prange


@njit(int32(int32, int32))
def somme(a, b):
    return a + b
```

* **Signatures de fonctions:** Spécifier différentes implémentations pour différents types d'arguments.

```python
@njit
def calcul(a, b):
    if isinstance(a, int32) and isinstance(b, int32):
        # Implémentation pour des entiers
        return a + b
    elif isinstance(a, float64) and isinstance(b, float64):
        # Implémentation pour des nombres à virgule flottante
        return a * b
```


## Compilation conditionnelle

* **`@njit(if_true=..., if_false=...)`:** Compiler différentes versions d'une fonction en fonction d'une condition.


```python
@njit(parallel=True, if_true=njit(parallel=False))
def calcul_parallele(n):
    resultat = 0
    for i in prange(n):
        resultat += i
    return resultat
```

* **`@generated_jit(nopython=True)`:** Générer dynamiquement le code de la fonction en fonction des types d'arguments.


## Accès à faible niveau

* **`cffi` et `ctypes`:** Interagir avec des bibliothèques C externes pour des performances optimales.


```python
from cffi import FFI

ffi = FFI()
ffi.cdef("double sin(double x);")
libm = ffi.dlopen("libm.so.6")

@njit
def sinus_c(x):
    return libm.sin(x)
```

## scipy.integrate : Intégration Numérique

* **`quad` :** Calcule l'intégrale définie d'une fonction univariée.
* **`dblquad`, `tplquad` :** Intégrales doubles et triples.
* **`odeint` :** Résout des équations différentielles ordinaires.
* **Méthodes avancées:**
    * Choix de l'algorithme d'intégration (par exemple, `quadrature`, `romberg`, etc.)
    * Gestion des singularités et des points problématiques.

***Exemple de base:***

```python
import scipy.integrate as integrate

def f(x):
    return x**2

resultat, erreur = integrate.quad(f, 0, 1)
print(f"Intégrale de x^2 de 0 à 1: {resultat:.4f}")

def g(x, y):
    return x*y

resultat, erreur = integrate.dblquad(g, 0, 1, lambda x: 0, lambda x: 1)
print(f"Intégrale double de x*y sur [0,1]x[0,1]: {resultat:.4f}")
```

## scipy.optimize : Optimisation

* **`minimize` :** Trouve le minimum d'une fonction scalaire.
* **`minimize_scalar` :** Minimisation univariée.
* **`root_scalar` :** Trouve la racine d'une fonction scalaire.
* **`fsolve` :** Résout un système d'équations non linéaires.
* **Méthodes avancé    es:**
* Choix de l'algorithme d'optimisation (par exemple, `BFGS`, `Nelder-Mead`,     etc.)
* Contraintes d'optimi    sation.
* Optimisation globale vs
 
***Exemple de base:***. locale.

```python
import scipy.optimize as optimize

def rosen(x):
    return sum(100.0*(x[1:]-x[:-1]**2.0)**2.0 + (1-x[:-1])**2.0)

resultat = optimize.minimize(rosen, x0=[1.3, 0.7, 0.8, 1.9, 1.2])
print(f"Minimum de la fonction de Rosenbrock: {resultat.x}")

def f(x):
    return x**2 - 2

sol = optimize.root_scalar(f, bracket=[0, 2])
print(f"Racine de x^2 - 2: {sol.root}")
```

## scipy.signal : Traitement du signal

* **Filtrage :** `lfilter`, `filtfilt`, etc. pour appliquer des filtres numériques aux signaux.
* **Transformée de Fourier :** `fft`, `ifft`, etc. pour l'analyse fréquentielle des signaux.

* **Méthodes avancé    es:**
* Conception de filtres numér    iques.
* Analyse temps-fréquence (wavelets, spectrog
 
***Exemple de base:***rammes).

```python
%matplotlib inline
import scipy.signal as signal
signal_bruite = np.random.normal(0, 1, 1000)  # Exemple de signal bruité

b, a = signal.butter(4, 0.2)  # Filtre passe-bas Butterworth d'ordre 4
signal_filtre = signal.lfilter(b, a, signal_bruite)  # Application du filtre

frequences, spectre_puissance = signal.periodogram(signal_filtre)  # Utilisation du signal filtré
plt.xlim(min(frequences),max(frequences))
plt.ylim(min(spectre_puissance),max(spectre_puissance))
plt.plot(frequences, spectre_puissance)
plt.xlabel("Fréquence")
plt.ylabel("Puissance")
plt.title("Spectre de puissance du signal filtré")
plt.show()
```

# SymPy 

SymPy est une bibliothèque Python puissante pour le calcul symbolique, permettant de manipuler des expressions mathématiques de manière symbolique plutôt que numérique. Cela signifie que vous pouvez travailler avec des variables, des équations, des fonctions, etc., sans avoir besoin de valeurs numériques spécifiques.


## Introduction à SymPy

### Symboles et expressions

* Créer des symboles : `x = sp.Symbol('x')`, `y = sp.Symbol('y')`
* Définir des expressions : `expr = x**2 + y`
* Afficher joliment : `sp.pprint(expr)`

```python
import sympy as sp
x, y = sp.symbols('x y')
expr = x**2 + y
sp.pprint(expr)
```

### Opérations de base

* Simplification : `sp.simplify(expr)`
* Factorisation : `sp.factor(expr)`
* Développement : `sp.expand(expr)`
* Substitution : `expr.subs(x, 2)`


```python
simplified_expr = sp.simplify(x**2 - 2*x + 1)
sp.pprint(simplified_expr)  # Affiche (x - 1)**2

factored_expr = sp.factor(x**2 - 1)
sp.pprint(factored_expr)    # Affiche (x - 1)*(x + 1)

expanded_expr = sp.expand((x + 1)**3)
sp.pprint(expanded_expr)    # Affiche x**3 + 3*x**2 + 3*x + 1

substituted_expr = expr.subs(x, 2)
sp.pprint(substituted_expr) # Affiche 4 + y
```

### Équations

* Créer des équations : `eq = sp.Eq(x**2 - 1, 0)`
* Résoudre des équations : `sp.solve(eq, x)`

```python
eq = sp.Eq(x**2 - 1, 0)
solutions = sp.solve(eq, x)
print(solutions)  # Affiche [-1, 1]
```

### Calcul différentiel et intégral

* Dérivées : `sp.diff(expr, x)`
* Intégrales : `sp.integrate(expr, x)`

```python
f = sp.sin(x)
derivee = sp.diff(f, x)
sp.pprint(derivee)   # Affiche cos(x)

integrale = sp.integrate(f, x)
sp.pprint(integrale) # Affiche -cos(x)
```


## Calcul Symbolique vs Numérique : Deux approches distinctes

Le monde des mathématiques utilise deux approches fondamentales pour manipuler et analyser les expressions : le calcul symbolique et le calcul numérique. Chacun a ses forces et ses faiblesses, et le choix de la meilleure approche dépend du problème à résoudre.

### Calcul Symbolique : La précision de l'abstraction

* **Manipulation de symboles:** Au lieu de travailler avec des valeurs numériques spécifiques, le calcul symbolique utilise des symboles (comme `x`, `y`, `a`, `b`) pour représenter des variables et des quantités inconnues.
* **Solutions exactes:** Il se concentre sur la manipulation de ces symboles selon les règles mathématiques pour obtenir des solutions exactes et analytiques, sans approximations ni erreurs d'arrondi.
* **Exemples d'applications:**
	* **Dérivation et intégration symbolique:** Trouver la dérivée ou l'intégrale d'une fonction sous forme symbolique, sans avoir à calculer des valeurs numériques.
	* **Résolution d'équations:** Trouver des solutions exactes à des équations algébriques ou différentielles.
	* **Simplification d'expressions:** Manipuler des expressions complexes pour les rendre plus simples et plus faciles à comprendre.

### Calcul Numérique : La puissance de l'approximation

* **Valeurs numériques:** Le calcul numérique traite des valeurs numériques spécifiques. Il utilise des algorithmes et des méthodes numériques pour obtenir des approximations des solutions.
* **Précision limitée:** En raison des limitations de la représentation des nombres en machine et des erreurs d'arrondi, les résultats du calcul numérique ne sont pas toujours exacts.
* **Exemples d'applications:**
	* **Simulation et modélisation:** Utiliser des modèles mathématiques pour simuler des systèmes complexes et prédire leur comportement.
	* **Optimisation:** Trouver les valeurs des paramètres qui minimisent ou maximisent une fonction.
	* **Apprentissage automatique:** Entraîner des modèles d'apprentissage automatique sur des données numériques pour faire des prédictions.

### Le choix de l'approche

* **Précision vs. efficacité:** Le calcul symbolique offre une précision absolue, mais peut être plus lent pour les problèmes complexes. Le calcul numérique est généralement plus rapide, mais au prix d'une précision limitée.
* **Nature du problème:** Si vous avez besoin d'une solution exacte ou si vous manipulez des expressions symboliques, le calcul symbolique est préférable. Si vous travaillez avec des données numériques et que vous pouvez tolérer une certaine approximation, le calcul numérique est plus adapté.

**En résumé, le calcul symbolique et le calcul numérique sont complémentaires. En comprenant leurs forces et leurs faiblesses, vous pouvez choisir la meilleure approche pour résoudre vos problèmes mathématiques et scientifiques.**
